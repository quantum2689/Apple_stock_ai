```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

**lode dataset**

```python
dataset = pd.read_csv("Apple Dataset.csv")
dataset
```

**split dataset into test and train**

```python
x_train = dataset[0:5477].drop('Date',axis=1)
x_train.drop('High',axis=1)
y_train = dataset['High'][0:5477]
x_test = dataset[5477:10954].drop('Date',axis=1)
x_test.drop('High',axis=1)
y_test = dataset['High'][5477:10954]
x_train
```

**into numpy array**

```python
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
x_train
```

**build the Ai model**

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(x_train.shape[1],)),
    tf.keras.layers.Dense(200,activation='relu'),
    tf.keras.layers.Dense(200,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

**compile the model**

```python
model.compile(optimizer='adam',
  loss='mean_squared_error',
  metrics=['accuracy'])
```

```python
model.fit(x_train,y_test,epochs=200)
```
